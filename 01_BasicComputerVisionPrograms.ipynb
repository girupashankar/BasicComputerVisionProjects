## Convert to grayscale img

```python
img = cv2.imread("atlas.jpg")
grayImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(grayImg)
```

## Save the GrayScaleImg

```python
cv2.imwrite("GrayScaleImage.jpg", grayImg)
grayImg = cv2.imread("GrayScaleImage.jpg")
cv2_imshow(grayImg)
```
